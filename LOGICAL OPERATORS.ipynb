```python
#Ques.3 = Take two boolean variables x = True and y = False. Print the result of x and y.
x = True
y = False
print(x and y)
print(x or y)
print(x == y)
print(x != y)
```

```python
#Ques.4 = Use the operator to check if either 5 > 10 or 2 < 8.
x = (5>10 or 2<8)
print(x)
```
