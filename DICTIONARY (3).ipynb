## 1.Create a dictionary and print it.

```python
# Create a dictionary that store the name and age of a person.
person = {"Name":"Alice","age":25} #Dict
print(person)
```

## 2. Access a value from a dictionary

```python
#Given the dictionary {"Name":"Bob","age":30}, print the age.
person = {"Name":"Bob","age":30} #Dict
print(person["age"])
```

## 3.Add a new key value pair to a dictionary.

```python
# Add a key "City" with value "New York" to the dictionary person.
person = {"Name":"Bob","age":30} #Dict
person["City"] = "New York"
print(person)
```

## 4.Updating an existing value in a dictionary.


```python
#Change the age of "Alice" to 28 in the given dictionary.
person = {"Name":"Alice","age":25} #Dict
person["age"] = 28
print(person)
```

## 5.Delete a key from a dictionary

```python
# Remove the "age" key from the dictionary.
person = {"Name":"Alice","age":25}
del person["age"]
print(person)
```

## 6. Get the length of dictionary

```python
# use to check the length of dictionary.
person = {"Name":"Alice","age":25}
print(len(person))
```

## 7. Use get() method to access value safely.


```python
# Access the "city" key using get(), provide default a'unknow" if not present.
person = {"Name":"Alice","age":25}
city = person.get("city","Unknown")
print(city)
```

## 8.Copy a dictionary.

```python
#create a copy of the dictionary person.
person = {"Name":"Alice","age":25}
copy_person = person.copy()
print(copy_person)
```
